```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        print(f"List starts as {arr}")
        key = arr[i]
        print(f"Insertion Value is: {key}")
    
        j = i-1
        print(f"Comparison value is {arr[j]}")
        
        while j >=0 and key < arr[j]:
            print(f"Insertion Value:{key} is less than Comparison Value:{arr[j]}")
            print("Copy Value")
            arr[j+1] = arr[j]
            print(f"New List:{arr}")
            j=j-1
        
        print("Insert New Value")
        arr[j+1] = key
        print(f"New Sorted List: {arr}")
        print()
            
```

```python
my_list = [3,2,5,1]

insertion_sort(my_list)
```
