```python
def bubble_sort(arr):
    n=len(arr)
    #number of passes
    for i in range(n):
        print(f"Pass:{i+1}")
        #looks at unsorted part
        for j in range(n - i - 1):
            print(arr)
            print(f"J:{j}")
            print(f"1st Element(J): {arr[j]}")
            print(f"2nd Element(J+1): {arr[j+1]}")
            
            #Swaps elments 
            if arr[j] < arr[j+1]:
                print("No Swap")
            elif arr[j] > arr[j+1]:
                print("Swap!")
                arr[j], arr[j+1] = arr[j+1], arr[j]
            print(f"New List: {arr}")
            print()
    return arr
```

```python
my_list = [3,1,2,20,4,8,18,4,5]

bubble_sort(my_list)
```
