```python
def min_calc(arr):
    
    smallest_value = float('inf')
    
    while len(arr)>1:
        print(f'List Length is: {len(arr)}')
        print(arr)
        if arr[0] < arr[1]:
            popped = arr.pop(0)
        else:
            popped = arr.pop(1)
        
        if popped < smallest_value:
            smallest_value = popped
        
        print(f"Smallest value thus far: {smallest_value}")
        print()
```

```python
my_list = [20,6,8,7,8,3,1]

min_calc(my_list)
```
