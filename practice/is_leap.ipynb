```python
def is_leap(year):
    leap = False
    if year%4==0:
        if year%100==0 and year%400==0:
            leap=True
        elif year%100!=0:
            leap=True
    return leap
```

```python
test_cases = [2000, 2400, 1800, 1900, 2100, 2200, 2300, 2500, 2016]
```

```python
for x in test_cases:
    print(f"{x}: Is Leap: {is_leap(x)}")
```
