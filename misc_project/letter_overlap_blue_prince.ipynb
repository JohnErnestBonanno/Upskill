```python
def letter_overlap(my_tuple):
    word1=my_tuple[0]
    word2=my_tuple[1]
    
    
    
    if len(word1) <= len(word2):
        word_short = word1.lower()
        word_long = word2.lower()
    else:
        word_short = word2.lower()
        word_long = word1.lower()
    
    my_string = f"{word_short.capitalize()} and {word_long.capitalize()} becomes: "
    
    for char in word_short:
        word_long = word_long.replace(char,"",1)

    return(my_string + word_long.upper())
```
