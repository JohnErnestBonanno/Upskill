```python
alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
def caesar_cipher (plaintext, shift):
    
    plaintext = plaintext.replace(" ","").lower()
    shift = shift%26
    
    code_phrase = [] 
    for letter in plaintext:
        position = (alphabet.index(letter)+shift)%26
        code_phrase.append(alphabet[position])
    code_phrase = "".join(str(letter) for letter in code_phrase)
    return code_phrase
```

```python
def decode_caesar(encrypted_text, shift):
    decrtyption_phrase = []
    for letter in encrypted_text:
        position = (alphabet.index(letter)-shift)%26
        decrtyption_phrase.append(alphabet[position])
    plaintext = "".join(str(letter) for letter in decrtyption_phrase)
    return plaintext
```

```python
#Brute Force
def crack_caesar(encrypted_text):
    decryption_phrase = []
    for i in range(0,26):
        decryption_phrase = []
        for letter in encrypted_text:
            position = (alphabet.index(letter)-i)%26
            decryption_phrase.append(alphabet[position])
        plaintext = "".join(str(letter) for letter in decryption_phrase)
        print(f"For {alphabet[i]}: Decodes to: {plaintext}")
```
