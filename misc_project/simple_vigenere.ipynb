```python
alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
def vigenere_cipher(plaintext, key):
    plaintext = plaintext.replace(" ","").lower()
    key = key.replace(" ","").lower()
    code_phrase = [] 
    for i in range(0,len(plaintext)):
        start_pos = alphabet.index(plaintext[i])
        extra_shift = alphabet.index(str(key[i%len(key)]))%26
        total_shift = (start_pos+extra_shift)%26
        code = alphabet[total_shift] 
        code_phrase.append(code)
    code_phrase = "".join(str(letter) for letter in code_phrase)
    return code_phrase
```

```python
def decode_vigenere(encrypted_text, key):
    encrypted_text = encrypted_text.replace(" ","").lower()
    key = key.replace(" ","").lower()
    decrypted_phrase = [] 
    for i in range(0,len(encrypted_text)):
        start_pos = alphabet.index(encrypted_text[i])
        extra_shift = alphabet.index(str(key[i%len(key)]))%26
        total_shift = (start_pos - extra_shift)%26
        decode = alphabet[total_shift] 
        decrypted_phrase.append(decode)
    decrypted_phrase = "".join(decrypted_phrase)
    return decrypted_phrase
```
