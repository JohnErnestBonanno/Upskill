```python
tests = [
[4, 7, 2, 8, 10, 9],
[10, 5, 40, 40.3],
[-5, -2, -1, -11],
[42],
[1000, 8],
[1000, 8, 9000],
[2, 5, 1, 1, 4]
]
```

```python
def max_value(nums):
    max_value=nums[0]
    for x in nums:
        if x>max_value:
            max_value=x
    return max_value
```

```python
for test in tests:
    (print(max_value(test)))
```
