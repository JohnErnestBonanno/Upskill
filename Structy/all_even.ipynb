```python
tests = [
[4, 90, 68, 6, -2],
[14, 40, 36, 3],
[30, 24, 2048, 0, 12, 50],
[7, 7, 7, 7],
[100],
[1, 2, 4, 6, 8],
[42, 18, 96, 4, 70, 12, 58, 30, 84, 26]
]
```

```python
def all_even(nums):
    all_even_flag = True
    for num in nums:
        if num % 2 == 1:
            all_even_flag = False
    return print(all_even_flag)
```

```python
for test in tests:
    all_even(test)
```
