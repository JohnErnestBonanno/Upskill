```python
def is_prime(n):
    
    is_prime=True
    
    #review list comprehension
    sum_of_digits = [int(digit) for digit in list(str(n))]
    
    trigger = (f"{n} is Prime")
    if n == 1:
        is_prime=True
    elif n in (2,3):
        is_prime=True
        trigger = ("1,2,3 are Prime")
    elif n%2==0:
        is_prime=False
        trigger = (f"{n} is divisible by 2")
    elif sum(sum_of_digits) % 3 ==0:
        is_prime=False
        trigger = (f"{n} is divisible by 3 - Sum of Digits = 3")
    else:
        for i in range(n-2,3,-2):
            if n%i==0:
                is_prime=False
                trigger = (f"{n} is divisible by {i}")
                break
    return n,is_prime, trigger
```
