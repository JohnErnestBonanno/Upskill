```python
tests = ["what a wonderful world",
         "have a nice day",
         "the quick brown fox jumped over the lazy dog",
         "who did eat the ham",
         "potato"
        ]
```

```python
def longest_word(sentence):
    word_list = sentence.split()
    
    len_holder = len(word_list[0])
    holder = word_list[0]
    
    for word in word_list[1:]:        
        if len(word) >= len_holder:
            holder=word
            len_holder = len(word) 
    return print(holder)
```

```python
for test in tests:
    longest_word(test)
```

```python
### improved:

def longest_word2(sentence):
    word_list = sentence.split()
    longest_word = ""    
    for word in word_list:        
        if len(word) >= len(longest_word):
            longest_word = word
    return print(longest_word)
```

```python
for test in tests:
    longest_word2(test)
```

Key Ideas
* using an empty string with a length of 0 avoids having to slice the string
* obtain the length of the string, rather than storing its value else where
