```python
import pandas as pd
import numpy as np

#R transfer
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
pandas2ri.activate()
```

```python
%%R
# Install once; comment out after the first run
#install.packages("hellodatascience", repos = "https://cloud.r-project.org")
#install.packages("skimr", repos = "https://cloud.r-project.org")

library(hellodatascience)
library(skimr)

data(planets)
head(planets)
```

```python
planets_python = robjects.r['planets']
# convert to pandas
planets_df = pandas2ri.rpy2py(planets_python)
```
